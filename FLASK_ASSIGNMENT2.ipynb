Q1. Explain GET and POST methods.

A1. The GET and POST methods are two HTTP (HyperText Transfer Protocol) request methods used to send data from a client (such as a web browser) to a server in order to retrieve or submit information. They are fundamental to web development and are used in various scenarios to facilitate communication between the client and server,by use of GET method we can send the  data but it will be visible to everyone in the url and if we use POST method then data will be transfered as a body and it will bw secured and not visible to anyone .


Q2. Why is request used in Flask?

A2.In Flask, the request object is used to access incoming request data when handling HTTP requests. It allows the developer to retrieve information sent by the client in the request, such as form data, URL parameters, headers, and other related information. 

Q3. Why is redirect() used in Flask?

A3. In Flask, the redirect() function is used to instruct the client's web browser to navigate to a different URL. This function is useful for directing users to another route or URL within the application, often as a response to a specific action or request.

Q4. What are templates in Flask? Why is the render_template() function used?

A4. In Flask, templates are a crucial component for building dynamic web applications by allowing developers to separate the structure of a web page (HTML) from its content and logic. Templates are typically written in a templating language (such as Jinja2 in the case of Flask), which allows for embedding variables, control flow, and other programming constructs within HTML.                                                                        The render_template() function in Flask is used to generate HTML pages dynamically by combining a template  with provided data, producing a customized web page to be sent as a response to the client's request.
