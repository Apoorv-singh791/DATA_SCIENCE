Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are the visual shapes that can be added to a plot to represent data. They include a wide range of shapes such as lines, circles, squares, wedges, rectangles, and more. These glyphs are used to visually represent the data points and relationships within the plot.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import output_file , show , figure
```

```python
x=[1,2,3]
y=[2,3,2]
p=figure(title='demo')
p.xaxis.axis_label='x-axis'
p.yaxis.axis_label='y-axis'
p.scatter(x,y,fill_color='red',legend_label='DANGER' ,size=20)
p.line(x,y,line_width=20,legend_label='blue_line')
p.square(x,y,size=12, color="yellow",legend_label='AREA')
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A4. the Bokeh server is a tool that helps create interactive data visualizations that can be updated in real time. It allows you to build web-based applications with live, dynamic plots.

To use the Bokeh server for real-time interactive plots:

Create an interactive plot: Design the plot using Bokeh's plotting interface, including data sources and interactive elements.

Run the Bokeh server: Start the Bokeh server with your interactive plot script. It hosts the plot and enables real-time updates.

Access the plot in a web browser: View the plot in a web browser, where you can interact with it using tools like pan, zoom, and hover.

Dynamic updates: The data in the plot can be updated dynamically based on user actions or external triggers, making the plot responsive and allowing real-time data visualization.

Embedding a Bokeh plot into a web page or dashboard can be achieved using popular web frameworks like Flask or Django. Flask:
Install the necessary packages:

Create a Flask application and configure a route for the page where the Bokeh plot will be embedded.

Generate the Bokeh plot as usual and save it as an HTML file.

In the Flask route, render the HTML file containing the Bokeh plot.

Run the Flask application to view the embedded Bokeh plot in the browser.

Generate the Bokeh plot and save it as an HTML file.
