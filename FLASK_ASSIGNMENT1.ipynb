Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask

apoorv = Flask(__name__)

@apoorv.route('/apoorv')
def apoor():
    return "Hello World!!"

if __name__ == "__main__":
    apoorv.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

A3. In Flask, app routing refers to the process of defining URL routes (or endpoints) within a web application. Each route is associated with a specific function that gets executed when a user accesses the corresponding URL. we use app routes because it allows us to map different URLs to specific pieces of code, enabling your application to respond to various client requests appropriately.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

A5. In Flask, the url_for() function is used for URL building. It generates a URL for the given endpoint (view function) by referencing the endpoint's name. This helps in maintaining a clean and consistent way to define URLs in your application, making it easier to manage changes in the URLs throughout your codebase

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the home page'

@app.route('/user/<username>')
def user_profile(username):
    return f'User Profile: {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print("URL for index():", url_for('index'))
        print("URL for user_profile():", url_for('user_profile', username='john'))
    app.run()
```
