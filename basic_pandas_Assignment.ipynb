Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

A6. A Series in pandas is a one-dimensional labeled array that can hold any data type. It can be modified by changing its values or adding/removing elements.

A DataFrame is a two-dimensional tabular data structure with labeled axes (rows and columns). It can be thought of as a dict-like container for Series objects. A DataFrame is mutable, meaning you can modify its values, add or remove columns, and perform various operations on it.

On the other hand, a Panel is a three-dimensional data structure in pandas that is less commonly used compared to Series and DataFrame                  So, series and dataframe is mutable in nature and panel is immutable.
